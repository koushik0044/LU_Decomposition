```python
import numpy as np
import pandas as pd
from numpy import zeros, argmax, array,random
from sympy import Matrix, init_printing
def swap_rows(A, i, j):
    A[[i, j]] = A[[j, i]]
    return A
```

```python
def LU_decomposition(A):
    n = len(A)
    L = np.eye(n)
    U = np.zeros((n, n))
    P = np.zeros((n , n))
    
    for i in range(0, n):
        for j in range(0, i):
            L[i, j] = (A[i, j] - np.dot(L[i, 0:j], U[0:j, j])) / U[j, j]
        for j in range(i, n):
            U[i, j] = A[i, j] - np.dot(L[i, 0:i], U[0:i, j])
             
    

    return (L, U)
```

```python
df = pd.read_csv('data.csv', sep=',', header=None).to_numpy()
print('\n ---------- DataSet ------------')
display(Matrix(df))

L,U = LU_decomposition(df)

print('\n ---------- Matrix L of the Dataset ------------')
display(Matrix(L))
print('\n ---------- Matrix U of the Dataset ------------')
display(Matrix(U))

B = np.eye(len(L))
b = np.array(B[:,0])

def forward_subs(L,b):
    y=[]
    for i in range(len(b)):
        y.append(b[i])
        for j in range(i):
            y[i]=y[i]-(L[i,j]*y[j])
        y[i]=y[i]/L[i,i]

    return y

def back_subs(U,y):
    x=np.zeros_like(y)
    for i in range(len(x),0,-1):
      x[i-1]=(y[i-1]-np.dot(U[i-1,i:],x[i:]))/U[i-1,i-1]

    return x

def solve_system_LU(L,U,b):
    y=forward_subs(L,b)
    x=back_subs(U,y)

    return x

n = len(L)
Result = np.zeros((n, n))
for i in range(0,n):
    b = np.array(B[:,i])
    Result[:,i] = solve_system_LU(L,U,b)
    
print('\n ---------- Inverse of the Dataset ------------')
display(Matrix(Result))
```
